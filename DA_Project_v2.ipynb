#### NOTE: Manhattan has about 20,000 Restaurants

###### SAMPLE: 1000 restaurants in "Manhattan"

```python
def get_restaurants(api_key,location):
    import requests
    data = []
    # Retrive business information from yelp
    # Change steps in offset 
    for offset in range(0,20,1):
        search_data = {
        'term': "restaurant",
        'location': location.replace(' ', '+'),
        'limit': 50,
        'offset':offset
        }
        url = API_HOST + SEARCH_PATH
        headers = {
            'Authorization': 'Bearer %s' % api_key,
        }
        response = requests.request('GET', url, headers=headers, params=search_data)
        if response.status_code == 200:
            data += response.json().get('businesses')
        elif response.status_code == 400:
            print('400 Bad Request')
            break
    return data
```

#### Clean-Up Resturant Categories

```python
# Clean-up the resturant 'categories' column
business['categories'] = business['categories'].apply(lambda x: [x[j]['alias'] for j in range(len(x))])
```

#### Find Restaurant of Same Type

```python
# 给一个restaurant type，找出相同type的resturant
# given the input restaurant type want to get, return only the restaurants that are same type
input_type = 'korean'
result = [input_type in business['categories'][i] for i in range(N)]
business_typed=business[result]
business_typed
```

#### Find Restaurant within the Given Distance

```python
def select_restaurant(df,distance=2000):
    df['distance']=df.get('distance').apply(lambda x:x<=distance)
    df = df[df['distance']==True]
    return df
```

```python
business_td=select_restaurant(business_typed)
```

#### Clean-Up Longitude & Latitude

```python
# Extract Longitude & Latitude Information
# New Columns
business_td['latitude'] = business_td['coordinates'].apply(lambda x:x['latitude'])
business_td['longitude'] = business_td['coordinates'].apply(lambda y:y['longitude'])
business_td.head()
```

#### Clean-up Address Information

```python
# Clean-up the resturant address
def addr(x):
    addr = ''
    for i in range(len(x['display_address'])-1):
        addr += x['display_address'][i] + ', '
    addr += x['display_address'][-1]
    return addr

# New Column
business_td['address'] = business_td['location'].apply(lambda x: addr(x))
business_td
```

#### Clean-up Transaction Column --> 3 columns T/F to indicate type of transactions a restaurante registered for

```python
# 3 types of Yelp transactions: pickup, delivery, restaurant_reservation.
business_td['pickup'] = business_td['transactions'].apply(lambda x: 'pickup' in x)
business_td['delivery'] = business_td['transactions'].apply(lambda x: 'delivery' in x)
business_td['restaurant_reservation'] = business_td['transactions'].apply(lambda x: 'restaurant_reservation' in x)

# IF we want to drop some columns
#business = business.drop(columns=['transactions'],axis=1)
print(f'{len(business_td.columns.values)} Fields included in dataset: {business_td.columns.values}')
```

###### SAMPLE: 100 restaurants  review (6 per each) in "Manhattan"

```python
def get_business_review(api_key,business_id):
    import json
    import requests
    business_path = BUSINESS_PATH + business_id+"/reviews"
    reviews =[]
    # Change steps in offset  
    for offset in range(0,100,100):
        search_data = {
        'offset':offset
        }
        url = API_HOST + business_path
        headers = {
            'Authorization': 'Bearer %s' % api_key,
        }
        
        response = requests.request('GET', url, headers=headers, params=search_data)
        
        if response.status_code == 200:
            reviews += response.json().get('reviews')
        elif response.status_code == 400:
            print('400 Bad Request')
            break
        
        review_text = ''
        for review in reviews:
            review_text += review['text']
    return review_text
```

```python
def get_reviews(restaurants):
    if not restaurants:
        return None
    review_list = list()
    for restaurant in restaurants:
        restaurant_name = restaurant['name']
        restaurant_id = restaurant['id']
        
        review_text = get_business_review(API_KEY,restaurant_id)
        
        review_list.append((restaurant_name,review_text))
    return review_list
        
```
